```python
input_size=256
dataset=[]
label=[]
from PIL import Image
```

```python
for i,image_name in enumerate(listno):
    if(image_name.split('.')[1]=="jpg" or image_name.split('.')[1]=="JPG"):
        image=cv2.imread('/kaggle/input/my-data/required data/'+'no/'+image_name)
        image=Image.fromarray(image,'RGB')
        image=image.resize((input_size,input_size))
        dataset.append(np.array(image))
        label.append(0)
```

```python
for i,image_name in enumerate(listyes):
    if(image_name.split('.')[1]=="jpg" or image_name.split('.')[1]=="JPG"):
        image=cv2.imread('/kaggle/input/my-data/required data/'+'yes/'+image_name)
        image=Image.fromarray(image,'RGB')
        image=image.resize((input_size,input_size))
        dataset.append(np.array(image))
        label.append(1)
```

```python
dataset=np.array(dataset)
label=np.array(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(dataset,label,test_size=0.2,random_state=0)
```

```python
num_batches = len(x_train) // batch_size
```

```python
from keras.utils import normalize
x_train=normalize(x_train,axis=1)
x_test=normalize(x_test,axis=1)
```

```python
model=Sequential()
```

```python
model.add(Conv2D(32,(3,3),input_shape=(input_size,input_size,3)))
```

```python
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32, (3, 3), kernel_initializer='he_uniform'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=16,verbose=1,epochs=10,validation_data=(x_test,y_test),shuffle=False)
```
